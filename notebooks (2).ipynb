```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

# Constants
radius = 1.0  # Radius of the electron's circular path
angular_velocity = 1.0  # Angular velocity (radians per frame)

# Create a figure and axis
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
electron, = ax.plot([], [], 'bo', markersize=5)

# Initialize the electron's position
def init():
    electron.set_data([], [])
    return electron,

# Update the electron's position in each frame
def update(frame):
    theta = angular_velocity * frame
    x = radius * np.cos(theta)
    y = radius * np.sin(theta)
    electron.set_data(x, y)
    return electron,

# Create an animation
ani = FuncAnimation(fig, update, frames=np.arange(0, 2 * np.pi, 0.05), init_func=init, blit=True)

# Display the animation
plt.title('Simplified Electron Motion')
plt.show()
```

```python
import pygame
from pygame.locals import *
from OpenGL.GL import *
from OpenGL.GLU import *
import random

class Particle:
    def __init__(self):
        self.position = [random.uniform(-1, 1), random.uniform(-1, 1), random.uniform(-1, 1)]
        self.rotation = [random.uniform(0, 360), random.uniform(0, 360), random.uniform(0, 360)]

    def update(self):
        for i in range(3):
            self.rotation[i] += ROTATION_SPEED

    def draw(self):
        glPushMatrix()
        glTranslatef(*self.position)
        glRotatef(self.rotation[0], 1, 0, 0)
        glRotatef(self.rotation[1], 0, 1, 0)
        glRotatef(self.rotation[2], 0, 0, 1)
        glColor3f(1, 0, 0)  # Red color
        glutSolidSphere(PARTICLE_RADIUS, 20, 20)
        glPopMatrix()

class ParticleSimulator:
    def __init__(self, num_particles, particle_radius, rotation_speed, screen_width, screen_height):
        self.NUM_PARTICLES = num_particles
        self.PARTICLE_RADIUS = particle_radius
        self.ROTATION_SPEED = rotation_speed
        self.SCREEN_WIDTH = screen_width
        self.SCREEN_HEIGHT = screen_height
        self.BACKGROUND_COLOR = (0, 0, 0)
        self.particles = [Particle() for _ in range(self.NUM_PARTICLES)]

    def initialize_pygame(self):
        pygame.init()
        display = (self.SCREEN_WIDTH, self.SCREEN_HEIGHT)
        pygame.display.set_mode(display, DOUBLEBUF | OPENGL)
        gluPerspective(45, (display[0] / display[1]), 0.1, 50.0)
        glTranslatef(0.0, 0.0, -5)


    def handle_events(self):
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                return False
        return True

    def update_particles(self):
        for particle in self.particles:
            particle.update()

    def clear_and_draw_particles(self):
        glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT)
        for particle in self.particles:
            particle.draw()

    def run(self):
        self.initialize_pygame()
        running = True

        while running:
            running = self.handle_events()

            self.update_particles()
            self.clear_and_draw_particles()

            pygame.display.flip()
            pygame.time.wait(10)

        pygame.quit()

if __name__ == "__main__":
    particle_simulator = ParticleSimulator(
        num_particles=5,
        particle_radius=0.1,
        rotation_speed=1,
        screen_width=800,
        screen_height=600
    )
    particle_simulator.run()
```
