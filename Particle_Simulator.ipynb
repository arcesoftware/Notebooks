```python
import pygame
import random
import math

# Initialize Pygame
pygame.init()

# Constants for the particle simulation
SCREEN_WIDTH = 1940
SCREEN_HEIGHT = 1280
PARTICLE_COUNT = 500
BACKGROUND_COLOR = (0, 0, 0)
MIN_DISTANCE_FROM_ORIGIN = 100  # Minimum distance from (0, 0)

# Precompute center and max distance
CENTER_X, CENTER_Y = SCREEN_WIDTH / 2, SCREEN_HEIGHT / 2
MAX_DISTANCE = math.sqrt(CENTER_X**2 + CENTER_Y**2)

# Function to generate gradient colors based on distance
def generate_color(distance):
    normalized_distance = distance / MAX_DISTANCE

    red = int(255 * (1 - normalized_distance))
    green = int(255 * normalized_distance)
    blue = int(255 * (1 - abs(normalized_distance - 0.5) * 2))

    return red, green, blue

# Create the screen
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Particle Simulation")

# Particle class
class Particle:
    def __init__(self):
        # Ensure particle starts away from origin by minimum distance
        while True:
            self.x = random.randint(0, SCREEN_WIDTH)
            self.y = random.randint(0, SCREEN_HEIGHT)
            distance_from_origin = math.sqrt(self.x ** 2 + self.y ** 2)
            if distance_from_origin >= MIN_DISTANCE_FROM_ORIGIN:
                break

        self.vx = random.uniform(-1, 1)
        self.vy = random.uniform(-1, 1)
        self.radius = 2
        self.initial_distance = math.sqrt((self.x - CENTER_X) ** 2 + (self.y - CENTER_Y) ** 2)

    def move(self, gravity_force):
        # Update position and wrap around edges
        self.x = (self.x + self.vx) % SCREEN_WIDTH
        self.y = (self.y + self.vy) % SCREEN_HEIGHT
        distance = math.sqrt((self.x - CENTER_X) ** 2 + (self.y - CENTER_Y) ** 2)

        # Get color based on distance
        self.color = generate_color(distance)

        # Calculate direction toward the center
        direction_x = (CENTER_X - self.x) / distance if distance != 0 else 0
        direction_y = (CENTER_Y - self.y) / distance if distance != 0 else 0

        # Apply gravitational pull
        self.vx += gravity_force * direction_x
        self.vy += gravity_force * direction_y

        # Scale radius by initial distance ratio
        self.radius = max(1, int(self.initial_distance / max(distance, 1)))

# Create particles
particles = [Particle() for _ in range(PARTICLE_COUNT)]

# Create a Clock object to control the frame rate
clock = pygame.time.Clock()

# Gravitational force
gravity_force = 0.2  # Adjust the strength of the force

# Main loop
running = True

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Clear the screen
    screen.fill(BACKGROUND_COLOR)

    # Draw all particles with gradient colors and gravitational force
    for particle in particles:
        particle.move(gravity_force)
        pygame.draw.circle(screen, particle.color, (int(particle.x), int(particle.y)), int(particle.radius))

    # Update the display
    pygame.display.update()

    # Control the frame rate
    clock.tick(60)  # Lowered frame rate

# Quit Pygame
pygame.quit()
```

```python
import pygame
import random
import math

# Initialize Pygame
pygame.init()

# Constants
SCREEN_WIDTH, SCREEN_HEIGHT = 1940, 1280
BACKGROUND_COLOR = (0, 0, 0)
GRAVITY_CONSTANT = 6.67430e-7  # Gravitational constant
CENTER_X, CENTER_Y = 0, 0  # Central mass at origin
CENTER_MASS = 1e10  # Large mass for central gravity
MAX_SATELLITES = 50  # Maximum number of satellites
SATELLITE_COUNT = 10  # Initial number of satellites

# Pygame setup
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Satellite Simulator")

# Satellite class
class Satellite:
    def __init__(self):
        # Randomly position satellite around the origin
        distance = random.uniform(100, 500)  # Random distance from center
        angle = random.uniform(0, 2 * math.pi)
        self.x = distance * math.cos(angle)
        self.y = distance * math.sin(angle)
        
        # Initialize with a velocity for circular or elliptical orbits
        speed = random.uniform(1, 3)  # Initial speed
        self.vx = -speed * math.sin(angle)  # Perpendicular to radius
        self.vy = speed * math.cos(angle)   # Fixed missing parenthesis here
        
        self.radius = 2
        self.color = (random.randint(100, 255), random.randint(100, 255), random.randint(100, 255))

    def apply_gravity(self):
        # Calculate gravitational pull toward the origin
        dx = CENTER_X - self.x
        dy = CENTER_Y - self.y
        distance = math.sqrt(dx ** 2 + dy ** 2) + 0.1  # Prevent division by zero

        # Gravitational force from the central mass
        force = GRAVITY_CONSTANT * CENTER_MASS / (distance ** 2)
        force_x = force * dx / distance
        force_y = force * dy / distance

        # Update velocity based on gravitational force
        self.vx += force_x
        self.vy += force_y

    def update_position(self):
        # Update position based on velocity (no screen wrapping)
        self.x += self.vx
        self.y += self.vy

    def draw(self, surface):
        # Offset for drawing relative to screen center
        draw_x = int(self.x + SCREEN_WIDTH / 2)
        draw_y = int(self.y + SCREEN_HEIGHT / 2)
        
        # Draw satellite on the screen
        pygame.draw.circle(surface, self.color, (draw_x, draw_y), self.radius)

# Create initial satellites
satellites = [Satellite() for _ in range(SATELLITE_COUNT)]

# Main loop
clock = pygame.time.Clock()
running = True

while running:
    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE and len(satellites) < MAX_SATELLITES:
                # Space bar to spawn a new satellite
                satellites.append(Satellite())

    # Clear the screen
    screen.fill(BACKGROUND_COLOR)

    # Update and draw each satellite
    for satellite in satellites:
        satellite.apply_gravity()  # Apply gravitational force
        satellite.update_position()  # Update position
        satellite.draw(screen)  # Draw satellite

    # Draw the central mass (optional visual representation)
    pygame.draw.circle(screen, (0, 0, 0), (SCREEN_WIDTH // 2, SCREEN_HEIGHT // 2), 15)  # Central mass as a yellow circle

    # Update display
    pygame.display.flip()

    # Control frame rate
    clock.tick(60)

# Quit Pygame
pygame.quit()
```

```python
import pygame
import random
import math

# Initialize Pygame
pygame.init()

# Constants
SCREEN_WIDTH, SCREEN_HEIGHT = 1940, 1280
BACKGROUND_COLOR = (0, 0, 0)
GRAVITY_CONSTANT = 6.67430e-7  # Gravitational constant
CENTER_MASS = 1e10  # Large mass for central gravity
MAX_SATELLITES = 500  # Maximum number of satellites
SATELLITE_COUNT = 100  # Initial number of satellites

# Pygame setup
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Satellite Simulator")

class Satellite:
    """Class representing a satellite in orbit."""
    
    def __init__(self):
        """Initialize the satellite with a random position and velocity."""
        distance = random.uniform(100, 500)  # Random distance from center
        angle = random.uniform(0, 2 * math.pi)
        self.x = distance * math.cos(angle)
        self.y = distance * math.sin(angle)
        
        speed = random.uniform(1, 3)  # Initial speed
        self.vx = -speed * math.sin(angle)  # Perpendicular to radius
        self.vy = speed * math.cos(angle)
        
        self.radius = 3
        self.color = (random.randint(100, 255), random.randint(100, 255), random.randint(100, 255))

    def apply_gravity(self):
        """Apply gravitational force toward the center."""
        dx = -self.x  # Central mass is at (0, 0)
        dy = -self.y
        distance = math.sqrt(dx ** 2 + dy ** 2) + 0.1  # Prevent division by zero
        
        force = GRAVITY_CONSTANT * CENTER_MASS / (distance ** 2)
        # Normalize the direction vector
        distance_vector_length = math.sqrt(dx**2 + dy**2)
        force_x = force * (dx / distance_vector_length)
        force_y = force * (dy / distance_vector_length)

        self.vx += force_x
        self.vy += force_y

    def update_position(self):
        """Update satellite's position based on its velocity."""
        self.x += self.vx
        self.y += self.vy

    def draw(self, surface):
        """Draw the satellite on the specified surface."""
        draw_x = int(self.x + SCREEN_WIDTH / 2)
        draw_y = int(self.y + SCREEN_HEIGHT / 2)
        pygame.draw.circle(surface, self.color, (draw_x, draw_y), self.radius)

def handle_events(satellites):
    """Handle user input events."""
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            return False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE and len(satellites) < MAX_SATELLITES:
                satellites.append(Satellite())
    return True

# Create initial satellites
satellites = [Satellite() for _ in range(SATELLITE_COUNT)]

# Main loop
clock = pygame.time.Clock()
running = True

while running:
    running = handle_events(satellites)  # Handle events

    # Clear the screen
    screen.fill(BACKGROUND_COLOR)

    # Update and draw each satellite
    for satellite in satellites:
        satellite.apply_gravity()
        satellite.update_position()
        satellite.draw(screen)

    # Draw the central mass (optional visual representation)
    pygame.draw.circle(screen, (0, 0, 0), (SCREEN_WIDTH // 2, SCREEN_HEIGHT // 2), 15)  # Central mass as a yellow circle

    # Update display
    pygame.display.flip()

    # Control frame rate
    clock.tick(520)

# Quit Pygame
pygame.quit()
```
